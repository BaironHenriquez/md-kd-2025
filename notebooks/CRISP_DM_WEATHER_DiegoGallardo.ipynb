## Paso 2: Comprensión de los Datos y Limpieza

En esta sección se realiza la limpieza avanzada de datos, lo cual incluye:

- Imputación de valores faltantes.
- Conversión de variables categóricas a numéricas (one-hot encoding o label encoding).
- Detección y tratamiento de outliers.
- Conversión adecuada de tipos de datos.

Este paso es crucial para asegurar la calidad de los datos antes de aplicar modelos estadísticos y de machine learning.

```python
import pandas as pd
import numpy as np

# Cargar el dataset desde archivo
df = pd.read_csv("/home/kedro_docker/data/01_raw/weatherAUS.csv")

# Copiar el dataset para limpieza
df_cleaned = df.copy()

# Convertir la columna Fecha a tipo datetime
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])

# Eliminar columnas con más del 30% de valores nulos
threshold = 0.3
df_cleaned = df_cleaned.loc[:, df_cleaned.isnull().mean() < threshold]

# Rellenar valores nulos en columnas numéricas con la mediana
num_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[num_cols] = df_cleaned[num_cols].fillna(df_cleaned[num_cols].median())

# Rellenar valores nulos en columnas categóricas con el valor más frecuente
cat_cols = df_cleaned.select_dtypes(include=['object']).columns
df_cleaned[cat_cols] = df_cleaned[cat_cols].fillna(df_cleaned[cat_cols].mode().iloc[0])

# Convertir variables categóricas a tipo category
for col in cat_cols:
    df_cleaned[col] = df_cleaned[col].astype('category')

# Convertir etiquetas objetivo a binario
df_cleaned['RainTomorrow'] = df_cleaned['RainTomorrow'].map({'No': 0, 'Yes': 1})
df_cleaned['RainToday'] = df_cleaned['RainToday'].map({'No': 0, 'Yes': 1})

print("Limpieza de datos completada. Dimensiones del dataset limpio:", df_cleaned.shape)
```


## Paso 2: Comprensión de los Datos y Limpieza

En esta sección se aplica una limpieza exhaustiva del dataset, incluyendo:

- Conversión de columnas de fecha.
- Imputación de valores faltantes.
- Codificación de variables categóricas.
- Conversión de tipos de datos.
- Eliminación de columnas con alto porcentaje de nulos.


```python

import pandas as pd
import numpy as np

# Cargar datos
df = pd.read_csv("/home/kedro_docker/data/01_raw/weatherAUS.csv")

# Convertir columna de fecha
df['Date'] = pd.to_datetime(df['Date'])

# Eliminar columnas con más del 30% de valores nulos
df = df.loc[:, df.isnull().mean() < 0.3]

# Imputar valores numéricos con la mediana
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Imputar categóricos con la moda
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

# Convertir columnas categóricas a tipo category
for col in cat_cols:
    df[col] = df[col].astype('category')

# Convertir etiquetas objetivo a binario
df['RainTomorrow'] = df['RainTomorrow'].map({'No': 0, 'Yes': 1})
df['RainToday'] = df['RainToday'].map({'No': 0, 'Yes': 1})

df_cleaned = df.copy()
print("Datos limpiados. Dimensiones:", df_cleaned.shape)
```


## Paso 3: Modelos Estadísticos Paramétricos

### 3.1 Regresión Lineal para Predecir `RISK_MM`

Este modelo permite analizar la relación entre las variables climáticas y la cantidad de lluvia registrada.


```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Variables predictoras
features = ['MinTemp', 'MaxTemp', 'Humidity3pm', 'Temp3pm', 'Pressure3pm']
X = df_cleaned[features]
y = df_cleaned['RISK_MM']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

linreg = LinearRegression()
linreg.fit(X_train, y_train)
preds = linreg.predict(X_test)

mse = mean_squared_error(y_test, preds)
print("MSE Regresión Lineal:", mse)
```


### 3.2 Regresión Logística para Predecir `RainTomorrow`

Modelo de clasificación binaria para predecir si lloverá mañana.


```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df_cleaned[features]
y = df_cleaned['RainTomorrow']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Precisión Regresión Logística:", acc)
```


## Paso 4: Métodos Bayesianos

### Clasificador Naive Bayes para la variable `RainTomorrow`


```python

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
y_nb_pred = nb.predict(X_test)

nb_acc = accuracy_score(y_test, y_nb_pred)
print("Precisión Naive Bayes:", nb_acc)
```


## Paso 5: Modelos de Árboles

### Árbol de Decisión y Random Forest


```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Árbol de decisión
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
tree_acc = accuracy_score(y_test, tree.predict(X_test))
print("Precisión Árbol de Decisión:", tree_acc)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_acc = accuracy_score(y_test, rf.predict(X_test))
print("Precisión Random Forest:", rf_acc)
```


## Paso 6: Almacenamiento de Datos Relevantes en PostgreSQL
